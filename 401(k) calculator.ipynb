```python
import pandas as pd
import numpy as np
```

```python
d = {'deposit_list':[], 'tribune_list':[], 'rate_list':[]}

# user will input salary, pct_contrib, years via a form
# you can change the numbers in this notebook to see what the result will be for testing purposes

salary = 30000 
pct_contrib = 4
years = 30
deposit = 0 # this is calculated on our end
tribune = 0 # this is calculated on our end
rate = 0.065  #the rate of return is fixed for now, it would be nice for the user to be able to adjust this later

if pct_contrib > 0 and pct_contrib <= 2:
    deposit = salary * (pct_contrib / 100)
    tribune = deposit

elif pct_contrib > 2 and pct_contrib <=6:
    tribune = (salary * ((pct_contrib - 2) / 100) * 0.5) + (salary * 0.02)
    deposit = salary * (pct_contrib / 100)

elif pct_contrib > 6:
    tribune = ((salary * 0.04) * 0.5) + (salary * 0.02)
    deposit = salary * (pct_contrib / 100)    
    
elif pct_contrib == 0:
    deposit = 0
    tribune = 0

else:
    print('no')
```

```python
deposit_list = []
tribune_list = []
rate_list = []


for i in range(0, years):
    deposit_list.append(deposit)
    tribune_list.append(tribune)
    rate_list.append(rate)
    
d['deposit_list'] = deposit_list
d['tribune_list'] = tribune_list
d['rate_list'] = rate_list
```

```python
df = pd.DataFrame(data=d)
```

```python
df.index = np.arange(1, len(df)+1)
```

```python
# https://stackoverflow.com/questions/41859311/cumsum-as-a-new-column-in-an-existing-pandas-data

for i in range(0, years):
    
    df['with_Trib_match'] = ((df['deposit_list'] + df['tribune_list']) * df['rate_list'].shift().add(1).cumprod().fillna(1)).cumsum()
    df['without_Trib_match'] = (df['deposit_list'] * df['rate_list'].shift().add(1).cumprod().fillna(1)).cumsum()
```

```python
#round numbers to two decimal places except for the rate of return
df.round({'deposit_list': 0, 'tribune_list': 0, 'rate_list': 3, 'with_Trib_match': 2, 'without_Trib_match': 2})
```
